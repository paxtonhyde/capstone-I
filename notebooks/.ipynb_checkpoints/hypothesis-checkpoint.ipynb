```python
import os
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
```

```python
SCRIPT_DIRECTORY = os.path.realpath("")
HOME_DIRECTORY = os.path.split(SCRIPT_DIRECTORY)[0]
DATA_DIRECTORY = os.path.join(HOME_DIRECTORY, "data")
IMAGE_DIRECTORY = os.path.join(HOME_DIRECTORY, "images")
```

```python
path = f"{DATA_DIRECTORY}/country_preferences.csv"
by_country = pd.read_csv(path)

path = f"{DATA_DIRECTORY}/country_cluster_map.csv"
cluster = pd.read_csv(path)

###
cluster = cluster[['ISO3', 'Cluster']].rename(columns={"Cluster":"cluster"})
by_country = cluster.merge(by_country, on="ISO3")

# by_country['q_intervention'] = by_country['p_intervention'] * by_country['n_intervention']

# grouped = by_country.groupby('cluster').agg({'ISO3':'count', 'q_intervention' : 'sum', 'n_intervention' : 'sum'})
# grouped["mean_p_intervention"] = grouped['q_intervention'] / grouped['n_intervention']
# grouped.head()
factors = [label.split("_")[1] for label in list(by_country.columns)[::2] if "_" in label]
```

```python
factor = 'intervention'
by_country = by_country.where(by_country[f'n_{factor}'] != 0).dropna()
by_country.cluster = by_country.cluster.astype(int)

fig, axes = plt.subplots(3, 1, figsize =(8,9))

for cluster in [0,1,2]:
    group = by_country[by_country["cluster"] == cluster]
    x = group[f"p_{factor}"]
    ax = axes[cluster]
    ax.hist(x, bins=20, density=True)
    ax.set_xlim(0, 1)
    ax.set_title(f"Cluster {cluster}")

fig.tight_layout(pad = 1)
plt.show()
```

```python
from scipy.stats import norm
plt.style.use('seaborn')
plt.rcParams['font.size'] = 16.0

factors = [label.split("_")[1] for label in list(by_country.columns)[::2] if "_" in label]
cluster_stats = pd.DataFrame({'cluster':[0, 1, 2], 'cluster_name': ['West', 'East', 'South'],\
                             'color':["r", "g", "b"] })

factor = "gender"
by_country = by_country.where(by_country[f'n_{factor}'] != 0).dropna()

fig, axes = plt.subplots(3, 1, figsize=(8,9))
fig2, ax2 = plt.subplots()

for cluster, nextcluster in zip([0, 1, 2], [1, 2, 0]):
    group = by_country[by_country["cluster"] == cluster]
    means = np.array(group[f"p_{factor}"])
    total_responses = np.sum(group[f"n_{factor}"])
    country_weights = np.array(group[f"n_{factor}"] / total_responses)

    mu, std = np.mean(means), np.std(means)

    ## plotting
    ax = axes[cluster]
    x = np.linspace(np.min(means) - .1, np.max(means) + .1, 100)
    ax.plot(x, norm.pdf(x, loc=mu, scale=std), 'r-', alpha=0.8,\
            label=f"normal distribution")
    ax.hist(means, density=True, label=f"density (n = {len(means)})")
    ax.set_yticklabels([])
    ax.set_xlabel(f"Probability of choosing default {factor}")
    ax.set_xlim(np.min(x) - .1, np.max(x) + .1)
    ax.set_title(f"{cluster_stats.cluster_name[cluster]}")
    ax.legend(loc='best')
    plt.savefig(f"{IMAGE_DIRECTORY}/{factor}_hists.png", dpi=120)

    x = np.linspace(0, 1, 100)
    ax2.plot(x, norm.pdf(x, loc=mu, scale=std), f'{cluster_stats.color[cluster]}-', alpha=0.8,\
            label=f"{cluster_stats.cluster_name[cluster]} (mean = {round(mu, 3)})")
    ax2.set_xlim(norm.ppf(.001, loc=mu, scale=std), norm.ppf(.999, loc=mu, scale=std))
    ax2.axvline(x=mu, linestyle="--", color='k')
    ax2.set_yticklabels([])
    ax2.set_title(f"{factor.capitalize()} by cluster")
    ax2.legend(loc='best')
    plt.savefig(f"{IMAGE_DIRECTORY}/{factor}_distributions.png", dpi=120)

fig.tight_layout(pad=1)
fig2.tight_layout(pad=1)
```

```python
import scipy.stats as stats
from tabulate import tabulate

cluster_compared_p = np.zeros((3,3))

for cluster, nextcluster in zip([0, 1, 2], [1, 2, 0]):
    b = (cluster + 1) % 3
    group = by_country[by_country["cluster"] == cluster]
    groupnext = by_country[by_country["cluster"] == nextcluster]
    means = np.array(group[f"p_{factor}"])
    meansnext = np.array(groupnext[f"p_{factor}"])
    total_responses = np.sum(group[f"n_{factor}"])
    total_responsesnext = np.sum(groupnext[f"n_{factor}"])
    country_weights = np.array(group[f"n_{factor}"] / total_responses)
    country_weightsnext = np.array(groupnext[f"n_{factor}"] / total_responsesnext)

    bootstrap = np.random.choice(means, size=len(means), p=country_weights)
    bootstrapnext = np.random.choice(meansnext, size=len(meansnext), p=country_weightsnext)

    t, p = stats.ttest_ind(means, meansnext, equal_var=False)
    cluster_compared_p[cluster][nextcluster] = round(p, 5)
    cluster_compared_p[nextcluster][cluster] = round(p, 5)
        
significance_matrix = pd.DataFrame(data=cluster_compared_p, \
                                   index=['West', 'East', 'South'], \
                                   columns=['West', 'East', 'South'])
table = tabulate(significance_matrix, tablefmt='github')
```

```python
significance_level = 0.10
significance_matrix = pd.DataFrame(index=['West', 'East', 'South'], columns=['West', 'East', 'South'])

for cluster, nextcluster in zip([0, 1, 2], [1, 2, 0]):
    for i, factor in enumerate(factors):
        name, namenext = cluster_stats['cluster_name'][cluster], cluster_stats['cluster_name'][nextcluster]
        significance_matrix[name][namenext] = np.min(cluster_compared_p[cluster][nextcluster])}
        significance_matrix[namenext][name] = np.min(cluster_compared_p[cluster][nextcluster])
        
        p = cluster_compared_p[cluster][nextcluster][i]
        if p < significance_level:
            pass
            
significance_matrix.head(3)
```
