```python
import os
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
```

```python
SCRIPT_DIRECTORY = os.path.realpath("")
HOME_DIRECTORY = os.path.split(SCRIPT_DIRECTORY)[0]
DATA_DIRECTORY = os.path.join(HOME_DIRECTORY, "data")
```

```python
path = f"{DATA_DIRECTORY}/country_preferences.csv"
by_country = pd.read_csv(path)

path = f"{DATA_DIRECTORY}/country_cluster_map.csv"
cluster = pd.read_csv(path)

###
cluster = cluster[['ISO3', 'Cluster']].rename(columns={"Cluster":"cluster"})
by_country = cluster.merge(by_country, on="ISO3")

# by_country['q_intervention'] = by_country['p_intervention'] * by_country['n_intervention']

# grouped = by_country.groupby('cluster').agg({'ISO3':'count', 'q_intervention' : 'sum', 'n_intervention' : 'sum'})
# grouped["mean_p_intervention"] = grouped['q_intervention'] / grouped['n_intervention']
# grouped.head()
factors = [label.split("_")[1] for label in list(by_country.columns)[::2] if "_" in label]
```

```python
factor = 'intervention'
by_country = by_country.where(by_country[f'n_{factor}'] != 0).dropna()
by_country.cluster = by_country.cluster.astype(int)

fig, axes = plt.subplots(3, 1, figsize =(8,9))

for cluster in [0,1,2]:
    group = by_country[by_country["cluster"] == cluster]
    x = group[f"p_{factor}"]
    ax = axes[cluster]
    ax.hist(x, bins=20, density=True)
    ax.set_xlim(0, 1)
    ax.set_title(f"Cluster {cluster}")

fig.tight_layout(pad = 1)
plt.show()
```
